Import Necessary Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Parsing imdb_top_1000 csv file into a data frame

```python
df = pd.read_csv("imdb_top_1000.csv")
```

Q1. What are the column names in the dataset?

```python
column = df.columns
print(*column, sep='\n')
```

```python
Q2. What is the average rating of the movies?
```

```python
avg_rating = df['IMDB_Rating'].mean()
avg_rating
```

Q3. How many unique genres are there?

```python
genre_count = df['Genre'].nunique()
genre_count
```

Q4. Save your results as a csv file

```python
my_file = pd.DataFrame({
    'Columns': column, 
    'Average Rating': avg_rating, 
    'Genre Count': genre_count
})
my_file.to_csv('mlt_onboarding.csv')
```

Q5. Plot a histogram of the genre distribution of the movies. How many movies from each genre are there in this list

```python
genre_count = df['Genre'].value_counts()

plt.bar(genre_count.index, genre_count.values)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Movie Genre Distribution')

plt.show()
```

Q6. Plot a scatterplot in this dataset for the number of movies vs year (Each year, how many movies made to this top 1000 list?)

```python
counts = df['Released_Year'].value_counts()

plt.scatter(counts.index, counts.values)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Year-Count Distribution')

plt.show()
```
